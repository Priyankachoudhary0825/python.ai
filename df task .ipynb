```python
import pandas as pd
#df=pd.read_excel('')
df=pd.read_csv('neet_data.csv')
df.head()
```

```python
print("Column Information:")
df.info()
print("\nTotal Rows and Columns:", df.shape)
```

```python
df_sorted=df.sort_values(by="retweet_count",ascending=False)
print("\n dataset sorted by retweet count:")
print(df_sorted.head())
```

```python
higest=df_sorted.iloc[0]
[['source','retweet_count','user_name','tweet','created_at','user_url']]
print("\nHigest Retweet Record:")
print(higest)
```

```python
top10=df_sorted.head(10)
top10_sum=top10['retweet_count'].sum()
print("\n TOP10 Retweet sum is",top10_sum)
```

```python
android_df=df[df['source'].str.contains('Android',na=False)]
print("\n Total tweets via Twitter for android:", android_df['retweet_count'].sum())
```

```python
higest_follow=df.loc[df['user_follower_count'].idxmax(),['user_name','user_location']]
print("\n User with Higher Followers:")
print(higest_follow)
```

```python
df['followers_friend_ratio']=df['user_follower_count']/df['user_friends_count']
print("\nFollowers_friends_ratio column created")
```

```python
print("\nNull values in each column:")
print(df.isnull().sum())
```

```python
df['user_description'].fillna("Twitter user",inplace=True)
print("\nNA values in user_description filled")
```

```python
df.dropna(axis=1,inplace=True)
print("\nColumns with null values dropped")
```

```python
df.insert(2,"is_allow",True)
print("\nColumn 'is_allow' added at 3rd position")
```

```python
if'tweet' in df.columns:
    df.drop('tweet',axis =1,inplace=True)
    print("\nTweet column dropped")
```

```python
df['created_at']=df['created_at'].astype(str)
month_year=df['created_at'].str.split("-").str[0:2]
print("\n Month and year from Created_at:")
print(month_year.head())
```

```python
print("\n Final DataFrame:")
print(df.head())
```
