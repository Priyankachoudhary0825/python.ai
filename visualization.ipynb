```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("zomato_dataset.csv")
print(df.head())
```

```python
df.columns
```

```python
sns.histplot(df['Delivery Rating'],bins=20)
plt.title("dining rating didtribution")
plt.xlabel("Delivery Rating")
plt.show()
```

```python
sns.scatterplot(x='Dining Votes',y='Dining Rating',data=df)
plt.title("Dining Votes vs Dining Rating")
plt.show()
```

```python
plt.hist(df['Prices'])
plt.title("Price Distribution")
plt.show()
```

```python
plt.scatter(df['Prices'] ,df['Votes'])
plt.xlabel("Prices")
plt.ylabel("Votes")
plt.show()
```

```python
df.City.value_counts()
```

```python
city_Labels=df.City.value_counts().index
city_Labels[:5]
```

```python
city_values=df.City.value_counts().values
city_values[:5]
```

```python
plt.pie(x=city_values[:5],labels=city_Labels[:5], autopct="%1.0f%%")
plt.show()
```
