```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text=[
    "Natural language processing is amazing","Machine learning makes NLP powerful",
    "I love learning NLP using Python"
]
```

```python
tokenized_text=[word_tokenize(sentence)
           for sentence in text  ]
print("Tokenized Text:")
print(tokenized_text)
```

```python
stop_words =set(stopwords.words('english'))
filtered_text=[]
for sentence in tokenized_text:
    filter_sentence=[word for word in sentence if word.lower() not in stop_words]
    filtered_text.append(filter_sentence)
    print("\nAfter Stopword Removal:")
    print(filtered_text)
```

```python
processed_text=[" ".join(sentence)for sentence in filtered_text]
```

```python
vectorizer=CountVectorizer()
x=vectorizer.fit_transform(processed_text)
print("\nFeature Names:")
print(vectorizer.get_feature_names_out())
print("\nVectorized output:")
print(x.toarray())
```
